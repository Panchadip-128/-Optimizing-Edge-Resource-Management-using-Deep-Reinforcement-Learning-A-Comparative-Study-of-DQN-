```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# =============================================================================
# 1. Simulation Parameters and Advanced Channel Model Settings
# =============================================================================

num_bits = 10000          # Number of bits per simulation run (raw key length)
raw_key_rate = 1e6        # Raw key generation rate (bits per second)
# Define a range of desired theoretical error probabilities (used as channel condition indicators)
desired_error_rates = np.linspace(0.01, 0.20, 20)

# Latency parameters (base latencies in seconds)
latency_base = 0.001      # Base latency for key exchange process
hardware_delay = 0.002    # Fixed hardware processing delay
# Additional latency factor for error correction (assumed to scale linearly with measured error)
latency_error_factor = 0.01  # 10ms extra per 100% error

# Error-correction and privacy amplification parameters
ec_efficiency_factor = 1.15   # Efficiency overhead for error correction (f-factor)
privacy_factor = 0.90         # Privacy amplification factor (fraction of key retained)

# Cost model parameters (example figures in USD)
hardware_cost_per_station = 50000   # CAPEX per QKD station
operational_cost_per_station = 1000   # Monthly OPEX per station
simulation_time = 60                  # Simulation duration in seconds
num_stations = 10                     # Number of QKD stations in the network

# =============================================================================
# 2. Utility Functions
# =============================================================================

def binary_entropy(e):
    """Calculate binary entropy (in bits). Handle edge cases."""
    if e <= 0 or e >= 1:
        return 0.0
    return -e * np.log2(e) - (1 - e) * np.log2(1 - e)

def compute_noise_variance(desired_error):
    """
    Given a target error probability for BPSK in Rayleigh fading,
    invert the theoretical BER formula to determine the required noise variance.
    For BPSK in Rayleigh fading, average BER is: 0.5 * (1 - sqrt(SNR/(1+SNR)))
    Solve for SNR, then noise_variance = 1/SNR (assuming signal amplitude = 1).
    """
    # Ensure desired_error is in valid range (< 0.5)
    if desired_error >= 0.5:
        desired_error = 0.499
    term = 1 - 2 * desired_error  # = sqrt(SNR/(1+SNR))
    snr = (term ** 2) / (1 - term ** 2)
    noise_var = 1 / snr
    return noise_var

# =============================================================================
# 3. Advanced QKD Channel Simulation
# =============================================================================

def simulate_qkd_advanced(num_bits, desired_error):
    """
    Simulate a QKD transmission over an advanced channel model:
      - Use BPSK modulation with values mapped to {-1, +1}
      - Apply Rayleigh fading (multiplicative channel gain)
      - Add AWGN noise with variance set to yield the desired error probability (on average)
    Returns:
      original_bits: array of transmitted bits in {-1, +1}
      received_bits: array of detected bits in {-1, +1}
      measured_error_rate: fraction of bits received in error
    """
    # Map bits to BPSK symbols (-1 and +1)
    original_bits = 2 * np.random.randint(0, 2, num_bits) - 1

    # For each bit, simulate a Rayleigh fading coefficient (scale parameter = 1)
    fading_coeffs = np.random.rayleigh(scale=1.0, size=num_bits)

    # Calculate noise variance based on desired error probability
    noise_var = compute_noise_variance(desired_error)
    noise_std = np.sqrt(noise_var)
    noise = np.random.normal(0, noise_std, num_bits)

    # Channel model: Received signal = fading_coefficient * transmitted_bit + noise
    received_signal = fading_coeffs * original_bits + noise

    # Decision: threshold at 0. If signal >= 0, decide +1; otherwise, decide -1
    received_bits = np.where(received_signal >= 0, 1, -1)

    # Compute measured error rate
    errors = np.sum(original_bits != received_bits)
    measured_error_rate = errors / num_bits

    return original_bits, received_bits, measured_error_rate

# =============================================================================
# 4. Advanced Secure Key Rate Calculation (with Error Correction & Privacy Amplification)
# =============================================================================

def compute_secure_key_rate_advanced(raw_rate, measured_error, ec_factor=ec_efficiency_factor, p_factor=privacy_factor):
    """
    Compute the final secure key rate after error correction and privacy amplification.
    Model:
      - Use the binary entropy function to account for the overhead of error correction.
      - The effective fraction of raw key retained is (1 - ec_factor * H2(e)).
      - Further reduce the key by the privacy amplification factor.
    """
    overhead = ec_factor * binary_entropy(measured_error)
    # Ensure that the overhead does not exceed 1 (i.e., complete key loss)
    effective_fraction = max(0, 1 - overhead)
    secure_rate = raw_rate * effective_fraction * p_factor
    return secure_rate

# =============================================================================
# 5. Latency and Enhanced Cost Model Functions
# =============================================================================

def compute_latency(measured_error):
    """
    Compute total latency as a function of measured error.
    Increased error requires more error correction processing.
    """
    additional_latency = measured_error * latency_error_factor
    return latency_base + hardware_delay + additional_latency

def cost_per_secure_bit_advanced(secure_key_rate, hardware_cost, operational_cost, sim_duration, num_stations):
    """
    Calculate the cost per secure bit over the simulation duration.
    The total cost includes CAPEX (hardware_cost per station) and prorated OPEX.
    """
    # Total CAPEX for all stations
    total_hardware_cost = num_stations * hardware_cost
    # Prorated operational cost for simulation duration (convert monthly cost to cost per second)
    total_operational_cost = num_stations * operational_cost * (sim_duration / (30 * 24 * 3600))
    total_cost = total_hardware_cost + total_operational_cost
    secure_bits_generated = secure_key_rate * sim_duration
    return total_cost / secure_bits_generated if secure_bits_generated > 0 else np.inf

# =============================================================================
# 6. Simulation & Data Collection with Sensitivity Analysis
# =============================================================================

results = []

for desired_error in desired_error_rates:
    # Advanced channel simulation with Rayleigh fading and AWGN
    _, _, measured_error = simulate_qkd_advanced(num_bits, desired_error)

    # Compute secure key rate using advanced error correction and privacy amplification model
    secure_rate = compute_secure_key_rate_advanced(raw_key_rate, measured_error)

    # Compute latency based on measured error rate
    latency = compute_latency(measured_error)

    # Compute cost per secure bit using enhanced cost model
    cost_bit = cost_per_secure_bit_advanced(secure_rate, hardware_cost_per_station,
                                            operational_cost_per_station, simulation_time, num_stations)

    results.append({
        "Desired_Error_Rate": desired_error,
        "Measured_Error_Rate": measured_error,
        "Secure_Key_Rate_bps": secure_rate,
        "Latency_sec": latency,
        "Cost_per_Secure_Bit_USD": cost_bit
    })

# Convert results to a DataFrame for analysis and visualization
df_results = pd.DataFrame(results)
print("Advanced Simulation Results:")
print(df_results)

# =============================================================================
# 7. Visualization of Advanced Simulation Results
# =============================================================================

plt.figure(figsize=(16, 5))

# (A) Secure Key Rate vs. Desired Error Probability
plt.subplot(1, 3, 1)
plt.plot(df_results["Desired_Error_Rate"], df_results["Secure_Key_Rate_bps"], marker='o', linestyle='-')
plt.xlabel("Desired Channel Error Probability")
plt.ylabel("Secure Key Rate (bps)")
plt.title("Secure Key Rate vs. Error Probability")
plt.grid(True)

# (B) Latency vs. Measured Error Rate
plt.subplot(1, 3, 2)
plt.plot(df_results["Measured_Error_Rate"], df_results["Latency_sec"], marker='s', color='orange', linestyle='-')
plt.xlabel("Measured Error Rate")
plt.ylabel("Latency (sec)")
plt.title("Latency vs. Measured Error Rate")
plt.grid(True)

# (C) Cost per Secure Bit vs. Measured Error Rate
plt.subplot(1, 3, 3)
plt.plot(df_results["Measured_Error_Rate"], df_results["Cost_per_Secure_Bit_USD"], marker='d', color='green', linestyle='-')
plt.xlabel("Measured Error Rate")
plt.ylabel("Cost per Secure Bit (USD)")
plt.title("Cost per Secure Bit vs. Measured Error Rate")
plt.grid(True)

plt.tight_layout()
plt.show()

# =============================================================================
# 8. Exporting Results for Documentation and Further Analysis
# =============================================================================

df_results.to_csv("advanced_qkd_simulation_results.csv", index=False)
print("\nAdvanced Simulation complete. Results exported to 'advanced_qkd_simulation_results.csv'.")
```

```python
import gym
from gym import spaces
import numpy as np
import random
import matplotlib.pyplot as plt

# =============================================================================
# 1. Custom Environment for Edge Resource Management
# =============================================================================

class EdgeResourceEnv(gym.Env):
    """
    A custom environment simulating edge resource management.
    - There are multiple edge nodes, each with a maximum capacity.
    - At each step, an incoming task with a random load must be assigned to one node.
    - Latency is modeled as a function of the node's load.
    - Loads decay over time to simulate task processing.
    """
    metadata = {'render.modes': ['human']}

    def __init__(self, num_nodes=3, capacity=100, decay_rate=0.1, base_latency=1.0):
        super(EdgeResourceEnv, self).__init__()
        self.num_nodes = num_nodes
        self.capacity = capacity          # Maximum load capacity per node
        self.decay_rate = decay_rate      # Fraction of load decayed per step
        self.base_latency = base_latency  # Base latency per task

        # Action space: which node (0 to num_nodes-1) to assign the incoming task
        self.action_space = spaces.Discrete(num_nodes)
        # Observation space: current load on each node, continuous between 0 and capacity
        self.observation_space = spaces.Box(low=0, high=capacity, shape=(num_nodes,), dtype=np.float32)

        self.reset()

    def reset(self):
        self.node_loads = np.zeros(self.num_nodes, dtype=np.float32)  # All nodes start with 0 load
        self.current_task = self._generate_task()  # Generate an initial task
        return self._get_state()

    def _generate_task(self):
        # Task load is a random integer between 5 and 20 units.
        return random.randint(5, 20)

    def _get_state(self):
        # Return a copy of the current node loads.
        return np.copy(self.node_loads)

    def step(self, action):
        # Apply decay to simulate processing (reduce load on each node)
        self.node_loads = self.node_loads * (1 - self.decay_rate)

        # Get the current task load
        task_load = self.current_task

        # Compute latency for the selected node:
        # latency = base_latency + (current_load + task_load) / capacity.
        latency = self.base_latency + (self.node_loads[action] + task_load) / self.capacity

        # Reward is defined as the negative latency (we want to minimize latency)
        reward = -latency

        # Update the chosen node's load by adding the task load
        self.node_loads[action] += task_load

        # Generate a new task for the next step
        self.current_task = self._generate_task()

        # This environment runs continuously (done flag remains False)
        done = False
        return self._get_state(), reward, done, {"latency": latency, "task_load": task_load}

    def render(self, mode='human'):
        print(f"Node loads: {self.node_loads}")

# =============================================================================
# 2. Discretization Function for Q-Learning
# =============================================================================

def discretize_state(state, bins=10, capacity=100):
    """
    Discretize the continuous state vector into a tuple of integers.
    Each node's load is mapped into a discrete bin.
    """
    discretized = tuple((state / capacity * (bins - 1)).astype(int))
    return discretized

# =============================================================================
# 3. Q-Learning Agent for Resource Management
# =============================================================================

# Q-learning hyperparameters
num_bins = 10
alpha = 0.1        # Learning rate
gamma = 0.95       # Discount factor
epsilon = 1.0      # Initial exploration rate
epsilon_min = 0.01
epsilon_decay = 0.995
episodes = 5000

env = EdgeResourceEnv()

# Initialize Q-table as a dictionary. The key is a discretized state tuple.
q_table = {}

def get_q(state):
    if state not in q_table:
        q_table[state] = np.zeros(env.num_nodes)
    return q_table[state]

# =============================================================================
# 4. Training Loop
# =============================================================================

rewards_per_episode = []

for episode in range(episodes):
    state = discretize_state(env.reset(), bins=num_bins, capacity=env.capacity)
    total_reward = 0

    # Each episode runs for 100 steps
    for step in range(100):
        # Epsilon-greedy action selection
        if np.random.rand() < epsilon:
            action = env.action_space.sample()
        else:
            q_values = get_q(state)
            action = np.argmax(q_values)

        next_state_cont, reward, done, info = env.step(action)
        next_state = discretize_state(next_state_cont, bins=num_bins, capacity=env.capacity)

        # Q-learning update rule
        best_next_q = np.max(get_q(next_state))
        current_q = get_q(state)[action]
        new_q = current_q + alpha * (reward + gamma * best_next_q - current_q)
        q_table[state][action] = new_q

        state = next_state
        total_reward += reward

        if done:
            break

    # Decay exploration rate
    epsilon = max(epsilon_min, epsilon * epsilon_decay)
    rewards_per_episode.append(total_reward)

    if episode % 500 == 0:
        print(f"Episode {episode}, Total Reward: {total_reward:.2f}, Epsilon: {epsilon:.3f}")

# Plot training rewards over episodes
plt.plot(rewards_per_episode)
plt.xlabel("Episode")
plt.ylabel("Total Reward")
plt.title("Training Rewards Over Episodes")
plt.show()

# =============================================================================
# 5. Testing the Trained Agent
# =============================================================================

test_episodes = 10
for ep in range(test_episodes):
    state = discretize_state(env.reset(), bins=num_bins, capacity=env.capacity)
    total_reward = 0
    print(f"\nTest Episode {ep+1}:")
    for step in range(20):
        q_values = get_q(state)
        action = np.argmax(q_values)
        next_state_cont, reward, done, info = env.step(action)
        print(f"Step {step+1}: Action: {action}, Reward: {reward:.2f}, Latency: {info['latency']:.2f}, Node Loads: {env.node_loads}")
        state = discretize_state(next_state_cont, bins=num_bins, capacity=env.capacity)
        total_reward += reward
    print(f"Total Reward in Test Episode {ep+1}: {total_reward:.2f}")
```

```python
import gym
from gym import spaces
import numpy as np
import random
import torch
import torch.nn as nn
import torch.optim as optim
from collections import deque
import matplotlib.pyplot as plt

# =============================================================================
# 1. Optimized Custom Environment for Edge Resource Management
# =============================================================================

class EdgeResourceEnv(gym.Env):
    """
    Environment simulating resource management over multiple edge nodes.
    Each node has a maximum capacity. At every step, a random task load arrives
    and must be assigned to one node. The latency is modeled as the sum of a base
    latency and the node's current load (normalized by capacity). An overload penalty
    is applied if a node’s projected load exceeds 90% of its capacity.
    """
    metadata = {'render.modes': ['human']}

    def __init__(self, num_nodes=3, capacity=100.0, decay_rate=0.1, base_latency=1.0):
        super(EdgeResourceEnv, self).__init__()
        self.num_nodes = num_nodes
        self.capacity = capacity          # Maximum load capacity per node
        self.decay_rate = decay_rate      # Fraction of load decayed per step
        self.base_latency = base_latency  # Base latency per task
        self.overload_threshold = 0.9 * self.capacity  # Overload penalty threshold

        # Action: assign task to one of the nodes
        self.action_space = spaces.Discrete(num_nodes)
        # Observation: current load on each node (normalized between 0 and 1)
        self.observation_space = spaces.Box(low=0, high=1, shape=(num_nodes,), dtype=np.float32)
        self.reset()

    def reset(self):
        self.node_loads = np.zeros(self.num_nodes, dtype=np.float32)
        self.current_task = self._generate_task()
        return self._get_state()

    def _generate_task(self):
        # Task load between 5 and 20 units.
        return random.randint(5, 20)

    def _get_state(self):
        # Normalize node loads to [0, 1] for stability.
        return np.clip(self.node_loads / self.capacity, 0, 1)

    def step(self, action):
        # Simulate processing: decay node loads
        self.node_loads *= (1 - self.decay_rate)

        task_load = self.current_task

        # Calculate latency as base_latency plus load fraction after assignment
        projected_load = self.node_loads[action] + task_load
        latency = self.base_latency + projected_load / self.capacity

        # Apply overload penalty if projected load exceeds threshold
        overload_penalty = 0.0
        if projected_load > self.overload_threshold:
            overload_penalty = - (projected_load - self.overload_threshold) / self.capacity

        # Reward is negative latency plus overload penalty (we want to minimize latency)
        reward = -latency + overload_penalty

        # Update the selected node's load
        self.node_loads[action] += task_load
        self.current_task = self._generate_task()

        done = False  # Continuing task
        info = {"latency": latency, "task_load": task_load}
        return self._get_state(), reward, done, info

    def render(self, mode='human'):
        print(f"Node loads: {self.node_loads}")

# =============================================================================
# 2. Double DQN Network Using PyTorch
# =============================================================================

class DQN(nn.Module):
    def __init__(self, state_size, action_size):
        super(DQN, self).__init__()
        self.fc1 = nn.Linear(state_size, 64)
        self.fc2 = nn.Linear(64, 64)
        self.fc3 = nn.Linear(64, action_size)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        return self.fc3(x)

# =============================================================================
# 3. Replay Buffer for Experience Replay
# =============================================================================

class ReplayBuffer:
    def __init__(self, capacity=10000):
        self.buffer = deque(maxlen=capacity)

    def push(self, state, action, reward, next_state, done):
        self.buffer.append((state, action, reward, next_state, done))

    def sample(self, batch_size):
        batch = random.sample(self.buffer, batch_size)
        state, action, reward, next_state, done = map(np.array, zip(*batch))
        return state, action, reward, next_state, done

    def __len__(self):
        return len(self.buffer)

# =============================================================================
# 4. Hyperparameters and Setup
# =============================================================================

env = EdgeResourceEnv(num_nodes=3, capacity=100.0, decay_rate=0.1, base_latency=1.0)
state_size = env.observation_space.shape[0]
action_size = env.action_space.n

# Hyperparameters
batch_size = 64
gamma = 0.99
lr = 0.001
target_update_freq = 10  # episodes
num_episodes = 500
max_steps = 100

# Epsilon parameters for exploration
epsilon = 1.0
epsilon_min = 0.01
epsilon_decay = 0.995

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Initialize policy and target networks
policy_net = DQN(state_size, action_size).to(device)
target_net = DQN(state_size, action_size).to(device)
target_net.load_state_dict(policy_net.state_dict())
target_net.eval()

optimizer = optim.Adam(policy_net.parameters(), lr=lr)
replay_buffer = ReplayBuffer(capacity=10000)

# =============================================================================
# 5. Training Loop with Double DQN Update
# =============================================================================

all_rewards = []
losses = []

for episode in range(num_episodes):
    state = env.reset()
    total_reward = 0
    for step in range(max_steps):
        # Epsilon-greedy action selection
        if random.random() < epsilon:
            action = env.action_space.sample()
        else:
            state_tensor = torch.FloatTensor(state).unsqueeze(0).to(device)
            with torch.no_grad():
                q_values = policy_net(state_tensor)
            action = torch.argmax(q_values).item()

        next_state, reward, done, info = env.step(action)
        total_reward += reward
        replay_buffer.push(state, action, reward, next_state, done)
        state = next_state

        # Only train if enough samples in the buffer
        if len(replay_buffer) >= batch_size:
            states, actions, rewards, next_states, dones = replay_buffer.sample(batch_size)

            states_tensor = torch.FloatTensor(states).to(device)
            actions_tensor = torch.LongTensor(actions).unsqueeze(1).to(device)
            rewards_tensor = torch.FloatTensor(rewards).unsqueeze(1).to(device)
            next_states_tensor = torch.FloatTensor(next_states).to(device)
            dones_tensor = torch.FloatTensor(dones).unsqueeze(1).to(device)

            # Current Q-values from policy network
            current_q = policy_net(states_tensor).gather(1, actions_tensor)
            # Double DQN: use policy_net to select best action, target_net to evaluate it
            next_actions = torch.argmax(policy_net(next_states_tensor), dim=1, keepdim=True)
            next_q = target_net(next_states_tensor).gather(1, next_actions)
            expected_q = rewards_tensor + gamma * next_q * (1 - dones_tensor)

            loss = nn.MSELoss()(current_q, expected_q)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            losses.append(loss.item())

    epsilon = max(epsilon_min, epsilon * epsilon_decay)
    all_rewards.append(total_reward)

    # Update target network periodically
    if episode % target_update_freq == 0:
        target_net.load_state_dict(policy_net.state_dict())

    if episode % 50 == 0:
        print(f"Episode {episode}, Total Reward: {total_reward:.2f}, Epsilon: {epsilon:.3f}")

# Plot training rewards and loss
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(all_rewards)
plt.xlabel("Episode")
plt.ylabel("Total Reward")
plt.title("Total Reward per Episode")

plt.subplot(1, 2, 2)
plt.plot(losses)
plt.xlabel("Training Steps")
plt.ylabel("Loss")
plt.title("Training Loss")
plt.tight_layout()
plt.show()

# =============================================================================
# 6. Testing the Trained Agent
# =============================================================================

test_episodes = 5
for ep in range(test_episodes):
    state = env.reset()
    total_reward = 0
    print(f"\nTest Episode {ep+1}:")
    for step in range(20):
        state_tensor = torch.FloatTensor(state).unsqueeze(0).to(device)
        with torch.no_grad():
            q_values = policy_net(state_tensor)
        action = torch.argmax(q_values).item()
        next_state, reward, done, info = env.step(action)
        total_reward += reward
        print(f"Step {step+1}: Action: {action}, Reward: {reward:.2f}, Latency: {info['latency']:.2f}, Node Loads: {env.node_loads}")
        state = next_state
    print(f"Total Reward in Test Episode {ep+1}: {total_reward:.2f}")
```

```python
import gym
from gym import spaces
import numpy as np
import random
import torch
import torch.nn as nn
import torch.optim as optim
from collections import deque
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import product

# =============================================================================
# 1. Custom Environment for Resource Management
# =============================================================================

class EdgeResourceEnv(gym.Env):
    """
    Simulates resource management over multiple edge nodes.
    Each node has a capacity, and tasks arrive with random load.
    Latency is a function of the node's load plus a base latency.
    An overload penalty is applied if the load exceeds 90% capacity.
    """
    metadata = {'render.modes': ['human']}

    def __init__(self, num_nodes=3, capacity=100.0, decay_rate=0.1, base_latency=1.0):
        super(EdgeResourceEnv, self).__init__()
        self.num_nodes = num_nodes
        self.capacity = capacity
        self.decay_rate = decay_rate
        self.base_latency = base_latency
        self.overload_threshold = 0.9 * self.capacity

        # Action: which node to assign the task
        self.action_space = spaces.Discrete(num_nodes)
        # Observation: normalized load on each node (0 to 1)
        self.observation_space = spaces.Box(low=0, high=1, shape=(num_nodes,), dtype=np.float32)

        self.reset()

    def reset(self):
        self.node_loads = np.zeros(self.num_nodes, dtype=np.float32)
        self.current_task = self._generate_task()
        return self._get_state()

    def _generate_task(self):
        # Task load between 5 and 20 units.
        return random.randint(5, 20)

    def _get_state(self):
        # Normalize node loads for stability.
        return np.clip(self.node_loads / self.capacity, 0, 1)

    def step(self, action):
        # Apply decay to simulate processing.
        self.node_loads *= (1 - self.decay_rate)

        task_load = self.current_task
        projected_load = self.node_loads[action] + task_load
        latency = self.base_latency + projected_load / self.capacity

        # Overload penalty if projected load exceeds threshold.
        overload_penalty = 0.0
        if projected_load > self.overload_threshold:
            overload_penalty = - (projected_load - self.overload_threshold) / self.capacity

        # Reward: negative latency (we wish to minimize latency)
        reward = -latency + overload_penalty

        self.node_loads[action] += task_load
        self.current_task = self._generate_task()

        done = False
        info = {"latency": latency}
        return self._get_state(), reward, done, info

    def render(self, mode='human'):
        print(f"Node loads: {self.node_loads}")

# =============================================================================
# 2. Double DQN Agent and Replay Buffer
# =============================================================================

class DQN(nn.Module):
    def __init__(self, state_size, action_size):
        super(DQN, self).__init__()
        self.fc1 = nn.Linear(state_size, 64)
        self.fc2 = nn.Linear(64, 64)
        self.fc3 = nn.Linear(64, action_size)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        return self.fc3(x)

class ReplayBuffer:
    def __init__(self, capacity=10000):
        self.buffer = deque(maxlen=capacity)

    def push(self, state, action, reward, next_state, done):
        self.buffer.append((state, action, reward, next_state, done))

    def sample(self, batch_size):
        batch = random.sample(self.buffer, batch_size)
        state, action, reward, next_state, done = map(np.array, zip(*batch))
        return state, action, reward, next_state, done

    def __len__(self):
        return len(self.buffer)

# =============================================================================
# 3. Training Function with Hyperparameters as Arguments
# =============================================================================

def train_dqn(env, episodes=50, steps=50, gamma=0.99, lr=0.001,
              batch_size=32, epsilon_start=1.0, epsilon_min=0.01,
              epsilon_decay=0.995, target_update_freq=5, device=None):
    """
    Trains a Double DQN agent on the environment.
    Returns the list of cumulative rewards per episode.
    """
    if device is None:
        device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    state_size = env.observation_space.shape[0]
    action_size = env.action_space.n

    policy_net = DQN(state_size, action_size).to(device)
    target_net = DQN(state_size, action_size).to(device)
    target_net.load_state_dict(policy_net.state_dict())
    target_net.eval()

    optimizer = optim.Adam(policy_net.parameters(), lr=lr)
    replay_buffer = ReplayBuffer(capacity=10000)

    epsilon = epsilon_start
    rewards_per_episode = []

    for ep in range(episodes):
        state = env.reset()
        total_reward = 0
        for st in range(steps):
            # Epsilon-greedy action selection.
            if random.random() < epsilon:
                action = env.action_space.sample()
            else:
                with torch.no_grad():
                    state_tensor = torch.FloatTensor(state).unsqueeze(0).to(device)
                    q_values = policy_net(state_tensor)
                action = q_values.argmax().item()

            next_state, reward, done, info = env.step(action)
            total_reward += reward

            replay_buffer.push(state, action, reward, next_state, done)
            state = next_state

            if len(replay_buffer) >= batch_size:
                states, actions, rewards, next_states, dones = replay_buffer.sample(batch_size)
                states_tensor = torch.FloatTensor(states).to(device)
                actions_tensor = torch.LongTensor(actions).unsqueeze(1).to(device)
                rewards_tensor = torch.FloatTensor(rewards).unsqueeze(1).to(device)
                next_states_tensor = torch.FloatTensor(next_states).to(device)
                dones_tensor = torch.FloatTensor(dones).unsqueeze(1).to(device)

                current_q = policy_net(states_tensor).gather(1, actions_tensor)
                next_actions = policy_net(next_states_tensor).argmax(dim=1, keepdim=True)
                next_q = target_net(next_states_tensor).gather(1, next_actions)

                expected_q = rewards_tensor + gamma * next_q * (1 - dones_tensor)
                loss = nn.MSELoss()(current_q, expected_q)

                optimizer.zero_grad()
                loss.backward()
                optimizer.step()

            if done:
                break

        rewards_per_episode.append(total_reward)
        epsilon = max(epsilon_min, epsilon * epsilon_decay)

        if ep % target_update_freq == 0:
            target_net.load_state_dict(policy_net.state_dict())

    return rewards_per_episode

# =============================================================================
# 4. Hyperparameter Tuning Routine (Grid Search)
# =============================================================================

def hyperparameter_tuning(env, param_grid, num_runs=3, episodes=50, steps=50):
    """
    Performs grid search over hyperparameters.
    param_grid: dict of hyperparameters and list of possible values.
    Returns a dictionary with hyperparameter combinations and their average final rewards.
    """
    best_params = None
    best_score = -np.inf  # Since rewards are negative, a higher (less negative) reward is better.
    results = []

    # Create combinations using itertools.product
    keys, values = zip(*param_grid.items())
    for combination in product(*values):
        params = dict(zip(keys, combination))
        final_rewards = []
        for run in range(num_runs):
            # Set random seeds for reproducibility
            seed = run
            random.seed(seed)
            np.random.seed(seed)
            torch.manual_seed(seed)

            # Create a fresh environment for each run.
            env_run = EdgeResourceEnv(num_nodes=3, capacity=100.0, decay_rate=0.1, base_latency=1.0)
            rewards = train_dqn(
                env_run,
                episodes=episodes,
                steps=steps,
                gamma=0.99,
                lr=params["lr"],
                batch_size=params["batch_size"],
                epsilon_start=1.0,
                epsilon_min=0.01,
                epsilon_decay=params["epsilon_decay"],
                target_update_freq=params["target_update_freq"]
            )
            final_rewards.append(rewards[-1])
        avg_final_reward = np.mean(final_rewards)
        results.append((params, avg_final_reward))
        print(f"Params: {params} -> Avg Final Reward: {avg_final_reward:.2f}")
        if avg_final_reward > best_score:
            best_score = avg_final_reward
            best_params = params

    return best_params, results

# Define hyperparameter grid
param_grid = {
    "lr": [0.001, 0.0005],
    "epsilon_decay": [0.995, 0.990],
    "batch_size": [32, 64],
    "target_update_freq": [5, 10]
}

# =============================================================================
# 5. Running Hyperparameter Tuning and Visualizing Results
# =============================================================================

if __name__ == "__main__":
    # Create a single environment instance (the tuning routine creates new envs as needed)
    env = EdgeResourceEnv(num_nodes=3, capacity=100.0, decay_rate=0.1, base_latency=1.0)

    best_params, tuning_results = hyperparameter_tuning(env, param_grid, num_runs=3, episodes=50, steps=50)
    print("\nBest Hyperparameters:", best_params)

    # Visualize tuning results: plot average final reward for each combination.
    # Create a DataFrame for visualization.
    tuning_df = []
    for params, score in tuning_results:
        row = params.copy()
        row["avg_final_reward"] = score
        tuning_df.append(row)
    tuning_df = pd.DataFrame(tuning_df)

    plt.figure(figsize=(10, 6))
    sns.barplot(x="lr", y="avg_final_reward", hue="epsilon_decay", data=tuning_df)
    plt.title("Hyperparameter Tuning: Avg Final Reward by lr and epsilon_decay")
    plt.ylabel("Avg Final Reward (Higher is Better)")
    plt.show()

    # You can create additional plots by grouping on batch_size or target_update_freq as needed.
```

```python
import gym
from gym import spaces
import numpy as np
import random
import torch
import torch.nn as nn
import torch.optim as optim
from collections import deque
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import product
from scipy.stats import ttest_ind

# =============================================================================
# 1. Custom Environment for Resource Management
# =============================================================================

class EdgeResourceEnv(gym.Env):
    """
    Simulates resource management over multiple edge nodes.
    Each node has a capacity, and tasks arrive with random load.
    Latency is a function of the node's load plus a base latency.
    An overload penalty is applied if the load exceeds 90% capacity.
    """
    metadata = {'render.modes': ['human']}

    def __init__(self, num_nodes=3, capacity=100.0, decay_rate=0.1, base_latency=1.0):
        super(EdgeResourceEnv, self).__init__()
        self.num_nodes = num_nodes
        self.capacity = capacity
        self.decay_rate = decay_rate
        self.base_latency = base_latency
        self.overload_threshold = 0.9 * self.capacity

        # Action: which node to assign the task
        self.action_space = spaces.Discrete(num_nodes)
        # Observation: normalized load on each node (0 to 1)
        self.observation_space = spaces.Box(low=0, high=1, shape=(num_nodes,), dtype=np.float32)

        self.reset()

    def reset(self):
        self.node_loads = np.zeros(self.num_nodes, dtype=np.float32)
        self.current_task = self._generate_task()
        return self._get_state()

    def _generate_task(self):
        # Task load between 5 and 20 units.
        return random.randint(5, 20)

    def _get_state(self):
        # Normalize node loads for stability.
        return np.clip(self.node_loads / self.capacity, 0, 1)

    def step(self, action):
        # Apply decay to simulate processing.
        self.node_loads *= (1 - self.decay_rate)

        task_load = self.current_task
        projected_load = self.node_loads[action] + task_load
        latency = self.base_latency + projected_load / self.capacity

        # Overload penalty if projected load exceeds threshold.
        overload_penalty = 0.0
        if projected_load > self.overload_threshold:
            overload_penalty = - (projected_load - self.overload_threshold) / self.capacity

        # Reward: negative latency plus overload penalty
        reward = -latency + overload_penalty

        self.node_loads[action] += task_load
        self.current_task = self._generate_task()

        done = False
        info = {"latency": latency}
        return self._get_state(), reward, done, info

    def render(self, mode='human'):
        print(f"Node loads: {self.node_loads}")

# =============================================================================
# 2. Baseline Policies
# =============================================================================

def baseline_naive(env, episodes=10, steps=50):
    rewards_per_episode = []
    for ep in range(episodes):
        state = env.reset()
        total_reward = 0
        for step_i in range(steps):
            action = step_i % env.num_nodes  # round-robin
            next_state, reward, done, info = env.step(action)
            total_reward += reward
            state = next_state
        rewards_per_episode.append(total_reward)
    return rewards_per_episode

def baseline_least_loaded(env, episodes=10, steps=50):
    rewards_per_episode = []
    for ep in range(episodes):
        state = env.reset()
        total_reward = 0
        for _ in range(steps):
            action = np.argmin(state)
            next_state, reward, done, info = env.step(action)
            total_reward += reward
            state = next_state
        rewards_per_episode.append(total_reward)
    return rewards_per_episode

# =============================================================================
# 3. Double DQN Agent and Replay Buffer
# =============================================================================

class DQN(nn.Module):
    def __init__(self, state_size, action_size):
        super(DQN, self).__init__()
        self.fc1 = nn.Linear(state_size, 64)
        self.fc2 = nn.Linear(64, 64)
        self.fc3 = nn.Linear(64, action_size)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        return self.fc3(x)

class ReplayBuffer:
    def __init__(self, capacity=10000):
        self.buffer = deque(maxlen=capacity)

    def push(self, state, action, reward, next_state, done):
        self.buffer.append((state, action, reward, next_state, done))

    def sample(self, batch_size):
        batch = random.sample(self.buffer, batch_size)
        state, action, reward, next_state, done = map(np.array, zip(*batch))
        return state, action, reward, next_state, done

    def __len__(self):
        return len(self.buffer)

def train_dqn(env, episodes=50, steps=50, gamma=0.99, lr=0.001,
              batch_size=32, epsilon_start=1.0, epsilon_min=0.01,
              epsilon_decay=0.995, target_update_freq=5, device=None):
    if device is None:
        device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    state_size = env.observation_space.shape[0]
    action_size = env.action_space.n

    policy_net = DQN(state_size, action_size).to(device)
    target_net = DQN(state_size, action_size).to(device)
    target_net.load_state_dict(policy_net.state_dict())
    target_net.eval()

    optimizer = optim.Adam(policy_net.parameters(), lr=lr)
    replay_buffer = ReplayBuffer(capacity=10000)

    epsilon = epsilon_start
    rewards_per_episode = []

    for ep in range(episodes):
        state = env.reset()
        total_reward = 0
        for st in range(steps):
            if random.random() < epsilon:
                action = env.action_space.sample()
            else:
                with torch.no_grad():
                    state_tensor = torch.FloatTensor(state).unsqueeze(0).to(device)
                    q_values = policy_net(state_tensor)
                action = q_values.argmax().item()

            next_state, reward, done, info = env.step(action)
            total_reward += reward
            replay_buffer.push(state, action, reward, next_state, done)
            state = next_state

            if len(replay_buffer) >= batch_size:
                states, actions, rewards, next_states, dones = replay_buffer.sample(batch_size)
                states_tensor = torch.FloatTensor(states).to(device)
                actions_tensor = torch.LongTensor(actions).unsqueeze(1).to(device)
                rewards_tensor = torch.FloatTensor(rewards).unsqueeze(1).to(device)
                next_states_tensor = torch.FloatTensor(next_states).to(device)
                dones_tensor = torch.FloatTensor(dones).unsqueeze(1).to(device)

                current_q = policy_net(states_tensor).gather(1, actions_tensor)
                next_actions = policy_net(next_states_tensor).argmax(dim=1, keepdim=True)
                next_q = target_net(next_states_tensor).gather(1, next_actions)
                expected_q = rewards_tensor + gamma * next_q * (1 - dones_tensor)

                loss = nn.MSELoss()(current_q, expected_q)
                optimizer.zero_grad()
                loss.backward()
                optimizer.step()

            if done:
                break

        rewards_per_episode.append(total_reward)
        epsilon = max(epsilon_min, epsilon * epsilon_decay)

        if ep % target_update_freq == 0:
            target_net.load_state_dict(policy_net.state_dict())

    return rewards_per_episode

# =============================================================================
# 4. Evaluation of Multiple Methods
# =============================================================================

# Best hyperparameters from tuning (from your previous results)
best_lr = 0.0005
best_epsilon_decay = 0.99
best_batch_size = 32
best_target_update_freq = 10

def evaluate_method(method, env, episodes=50, steps=50):
    if method == "round_robin":
        rewards = baseline_naive(env, episodes=episodes, steps=steps)
    elif method == "least_loaded":
        rewards = baseline_least_loaded(env, episodes=episodes, steps=steps)
    elif method == "dqn":
        rewards = train_dqn(
            env,
            episodes=episodes,
            steps=steps,
            gamma=0.99,
            lr=best_lr,
            batch_size=best_batch_size,
            epsilon_start=1.0,
            epsilon_min=0.01,
            epsilon_decay=best_epsilon_decay,
            target_update_freq=best_target_update_freq
        )
    elif method == "ppo":
        rewards = [0] * episodes  # Placeholder
    elif method == "a2c":
        rewards = [0] * episodes  # Placeholder
    else:
        rewards = []
    return rewards

methods = ["round_robin", "least_loaded", "dqn"]  # Add "ppo", "a2c" when implemented

def run_experiments(num_runs=3, episodes=50, steps=50):
    """
    Runs multiple seeds for each method and returns raw results (list of lists)
    without averaging them.
    """
    results = {m: [] for m in methods}

    for run in range(num_runs):
        random.seed(run)
        np.random.seed(run)
        torch.manual_seed(run)
        for m in methods:
            env = EdgeResourceEnv(num_nodes=3, capacity=100.0, decay_rate=0.1, base_latency=1.0)
            rewards = evaluate_method(m, env, episodes=episodes, steps=steps)
            results[m].append(rewards)

    return results

# =============================================================================
# 5. Run Experiments and Plot Results
# =============================================================================

if __name__ == "__main__":
    raw_results = run_experiments(num_runs=3, episodes=50, steps=50)

    # Average results over runs for visualization
    avg_results = {m: np.mean(raw_results[m], axis=0) for m in methods}

    plt.figure(figsize=(10, 6))
    for m, rewards in avg_results.items():
        plt.plot(rewards, label=m)
    plt.xlabel("Episode")
    plt.ylabel("Average Cumulative Reward")
    plt.title("Comparison of Methods Across Episodes")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Now, extract final-episode rewards for each run for statistical analysis.
    # Since raw_results[m] is a list of reward arrays (one per run),
    # we can get the final reward from each run.
    final_rewards_round_robin = np.array([r[-1] for r in raw_results["round_robin"]])
    final_rewards_dqn = np.array([r[-1] for r in raw_results["dqn"]])

    stat, p_val = ttest_ind(final_rewards_round_robin, final_rewards_dqn)
    print(f"T-test between Round Robin and DQN final-episode rewards: p={p_val:.4f}")
```
